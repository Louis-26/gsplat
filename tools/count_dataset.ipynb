```python
## RGBT-Scenes dataset image count
import os

DATASET_ROOT = "../RGB-T-dataset/RGBT-Scenes"
IMAGE_EXTS = {".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff"}

def is_image(filename: str) -> bool:
    return os.path.splitext(filename)[1].lower() in IMAGE_EXTS

def main():
    if not os.path.isdir(DATASET_ROOT):
        raise FileNotFoundError(f"Dataset folder not found: {DATASET_ROOT}")

    scenes = [
        d for d in os.listdir(DATASET_ROOT)
        if os.path.isdir(os.path.join(DATASET_ROOT, d))
    ]

    
    for scene in sorted(scenes):
        total = 0
        for mode in ["train", "test"]:
            scene_path = os.path.join(DATASET_ROOT, scene, "rgb", mode)
            count = 0
            for dirpath, _, filenames in os.walk(scene_path):
                for f in filenames:
                    if is_image(f):
                        count += 1
            total += count
            print(f"{scene} ({mode}): {count}")
        print(f"total image number of {scene}: {total}")
        print("=================================")



if __name__ == "__main__":
    main()
```
